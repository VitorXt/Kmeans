```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

file_path = 'iris.csv'
data = pd.read_csv(file_path, delimiter=';')

print(data.head())

X = data[['sepal_length', 'sepal_width']].values

k = 3

kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(X)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.figure(figsize=(10, 5))

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', alpha=0.6, edgecolor='k')

plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='X', s=200, label='Centros')

plt.title("K-Means Clustering do Dataset Iris")
plt.xlabel("Comprimento da Sépala (sepal_length)")
plt.ylabel("Largura da Sépala (sepal_width)")
plt.legend()
plt.show()

print("Número de clusters:", k)
print("Rótulos atribuídos pelo KMeans:", labels)
```
